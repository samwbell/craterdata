```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.gridspec as gridspec
from matplotlib.ticker import ScalarFormatter
import math
from scipy.stats.mstats import gmean
from scipy.stats import gamma, poisson, linregress, beta, norm, lognorm
from scipy import optimize
from scipy import interpolate
import time
import random
import pickle as pkl
from cratrcountr import *
%matplotlib inline
```

```python
fig = plt.figure(figsize=(13,2))

x_min, x_max = -1, 9

def customized_plotting_parameters(N, letter, x_min=x_min, x_max=x_max):
    plt.xlim([x_min, x_max])
    plt.ylim(-0.1, 3.3)
    plt.text(x_min, 1.2 + 1.1 * 2, letter, ha='left', va='top', size=20)
    plt.text(x_max, 1.0 * (1 + 1.1 * 2), 'N=' + str(N), ha='right', va='center', size=12)

def customized_PDF_comparison_plot(N, letter, x_min=-1, x_max=9, label_shift_x=0, 
                                   label_shift_y=0, sqrtN_label_shift_y=0):
    true_error_pdf_N = true_error_pdf(N).normalize()
    true_error_pdf_N.plot(color='mediumslateblue', ax=ax, label=True, 
                          xlim=[x_min, x_max], fig_size_adjustor=2,
                          label_shift_x=label_shift_x, label_shift_y=label_shift_y,
                          label_text_size=9)
    plt.text(x_max, 0.12 + 1.1 * 0, 'Poisson', size=7.5, color='mediumslateblue', ha='right')
    if N > 0:
        sqrt_N_error_pdf_N = sqrt_N_error_pdf(N).normalize()
        sqrt_N_error_pdf_N.plot(color='orange', ax=ax, upshift=1.1, label=True, 
                                xlim=[x_min, x_max], fig_size_adjustor=2,
                                label_shift_x=label_shift_x, 
                                label_shift_y=label_shift_y + sqrtN_label_shift_y,
                                label_text_size=9)
        plt.plot(sqrt_N_error_pdf_N.X, sqrt_N_error_pdf_N.P + 2.2, color='orange')
        plt.plot(true_error_pdf_N.X, true_error_pdf_N.P + 2.2, color='mediumslateblue')
        plt.text(x_max, 0.12 + 1.1 * 1, 'sqrt(N)', size=7.5, color='orange', ha='right')
    customized_plotting_parameters(N, letter, x_min=x_min, x_max=x_max)
    plt.xlabel('Cratering Rate Parameter ', labelpad=0, size=7.5)



def customized_error_bar_comparison_plot(N, letter, x_min=-1.0, x_max=9, label_shift_x=0, 
                                         label_shift_y=0, fig_size_adjustor=3.25,
                                         label_text_size=10, median_label_shift_y=0):
    true_error_pdf_N = true_error_pdf(N).normalize()
    true_error_pdf_N.plot(color='mediumslateblue', ax=ax, label=True, label_shift_x=label_shift_x,
                          label_shift_y=label_shift_y, fig_size_adjustor=fig_size_adjustor,
                          label_text_size=label_text_size)
    if N > 0:
        plot_log_of_normal_approximation(N, color='black', linewidth=0.7)
    true_error_pdf_N.plot(error_bar_type = 'linear', color='forestgreen', ax=ax, upshift=1.1,
                          label=True, label_shift_x=label_shift_x, label_shift_y=label_shift_y,
                          fig_size_adjustor=fig_size_adjustor, label_text_size=label_text_size)
    plot_linear_approximation(N, color='black', linewidth=0.7, upshift=1.1)
    true_error_pdf_N.plot(error_bar_type = 'median', color='crimson', ax=ax, upshift=2 * 1.1,
                          label=True, label_shift_x=label_shift_x, 
                          label_shift_y=label_shift_y + median_label_shift_y,
                          fig_size_adjustor=fig_size_adjustor, label_text_size=label_text_size)
    plot_median_approximation(N, color='black', linewidth=0.7, upshift=2.2)
    customized_plotting_parameters(N, letter, x_min=x_min, x_max=x_max)
    plt.text(x_max, 0.12 + 1.1 * 2, 'Median', size=7.5, color='crimson', ha='right')
    plt.text(x_max, 0.12 + 1.1 * 1, 'Linear', size=7.5, color='forestgreen', ha='right')
    plt.text(x_max, 0.12 + 1.1 * 0, 'Log', size=7.5, color='mediumslateblue', ha='right')
    plt.xlabel('Cratering Rate Parameter ', labelpad=0, size=7.5)
    

ax = fig.add_subplot(151)
customized_error_bar_comparison_plot(1, 'a', label_shift_x=-5.5, label_shift_y=-0.25, 
                                     fig_size_adjustor=1.5, label_text_size=9)

ax = fig.add_subplot(152)
customized_error_bar_comparison_plot(2, 'b', label_shift_x=-6.85, label_shift_y=-0.15, 
                                     fig_size_adjustor=1.5, label_text_size=9,
                                     median_label_shift_y=0.17)

ax = fig.add_subplot(153)
customized_error_bar_comparison_plot(5, 'c', label_shift_x=-7.9, label_shift_y=-0.04, 
                                     fig_size_adjustor=1.5, label_text_size=9, x_max=15,
                                     median_label_shift_y=0.1)

ax = fig.add_subplot(154)
customized_error_bar_comparison_plot(10, 'd', label_shift_x=-8.6, label_shift_y=-0.06, 
                                     fig_size_adjustor=1.5, label_text_size=9, 
                                     x_min=2, x_max=27, median_label_shift_y=0.06)

ax = fig.add_subplot(155)
customized_error_bar_comparison_plot(100, 'e', label_shift_x=53, label_shift_y=-0.25, 
                                     fig_size_adjustor=1.5, label_text_size=9,
                                     x_min=53, x_max=143)

plt.subplots_adjust(wspace=0.1, hspace=0.28)
pfig = plt.gcf()
pfig.savefig('../PDF_fit_explanation.pdf', bbox_inches='tight')
```
