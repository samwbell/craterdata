```python
for n_ps in [100, 200, 300, 400, 500, 1000, 10000, 100000]:
    t1 = time.time()
    print('Making figure for N = ' + str(n_ps) + '...')
    fig = plt.figure(figsize=(14.5, 6))
    ax = fig.add_subplot(111)
    plot_cumulative_unbinned(
        example_ds, area, plot_point_error_bars=False, 
        plot_points=False, plot_lines=False,
        do_formatting=True
    )
    xlim = [10**-0.05, 10**2.5]
    D = np.logspace(np.log10(xlim[0]), np.log10(xlim[1]), 10000)
    for i in range(n_ps):
        m, b = m_list[i], b_list[i]
        if i == 0:
            label = str(n_ps) + ' best fit lines'
        else:
            label = None
        plt.plot(
            D, 10**(m * np.log10(D) + b), color=cs[1], lw=0.1,
            label=label
        )
    plt.plot(
        D, 10**pf(np.log10(D)), color='black', lw=2,
        label='Model production function'
    )
    plt.legend(loc='lower left')
    plt.xlim(xlim)
    plt.ylim([2E-6, 2E-3])
    ax_inset = fig.add_axes([0.537, 0.545, 0.363, 0.332])
    slope_rv = make_pdf_from_samples(
        m_list[:n_ps], slope_data=True, 
        kind='mean'
    )
    erase_box(ax_inset)
    slope_rv.plot(
        label=True, rounding_n=3, label_text_size=25,
        label_shift_y=0.5, xlim=[-3.2, -1.1],
        pdf_label=str(n_ps) + ' slopes'
    )
    pfig = plt.gcf()
    file_base = 'figs/example_ds_plot_PDF_'
    if n_ps < 2000: 
        pfig.savefig(
            file_base + str(n_ps) + '.pdf', 
            bbox_inches='tight'
        )
    else:
        pfig.savefig(
            file_base + str(n_ps) + '.jpg', 
            bbox_inches='tight', dpi=150
        )
    t2 = time.time()
    print(format_runtime(t2 - t1))
```
