```python
def mle_bias_plot(mle_alphas, N, xlim=None, n_bins=1000):
    mean_str = str(round(np.mean(mle_alphas), 5))
    median_str = str(round(np.median(mle_alphas), 5))
    alpha_pdf = make_pdf_from_samples(
        mle_alphas, kind='log', n_bins=n_bins
    )
    (-1 * alpha_pdf).plot(
        label=True, rounding_n=3, 
        pdf_label=(
            'N = ' + str(N) + '\n'
            'Mean : ' + mean_str + '\n'
            'Median : ' + median_str + '\n'
        ),
        pdf_gap_shift = 0.3, xlim=xlim
    )

def pareto_T_dist(T, a, N):
    return a**N * (T)**(N - 1) * np.exp(-1 * a * T) / gf(N)
def pareto_mle_dist(mle, a, N):
    num = (a * N)**N * np.exp(-1 * a * N / mle)
    denom = mle**(N+1) * gf(N)
    return num / denom
```

```python
def reverse_fit_probs(
    observed_slope, mean_slope_factor, mean_upper, mean_lower,
    min_slope=0.2, max_slope=17.0, model_slope_n=1000,
    observed_slope_n=10000
):
    X_min = min_slope
    X_max = max_slope
    X = np.logspace(
        np.log10(X_min), np.log10(X_max), observed_slope_n, endpoint=False
    )
    slope_array = -1 * np.linspace(
        X_min, X_max, model_slope_n, endpoint=False
    )
    prob_list = []
    for test_slope in slope_array:
        log_max = np.log10(mean_slope_factor * -1 * test_slope)
        P = piecewise_normal_fit_equation(
            np.log10(X), log_max, mean_lower, mean_upper
        )
        P = P / trapezoid(X, P)
        prob = np.interp(-1 * observed_slope, X, P)
        prob_list.append(prob)
    return -1 * slope_array, -1 * np.array(prob_list)


def reverse_model(
    observed_slope, mean_slope_factor, mean_upper, mean_lower,
    min_slope=0.2, max_slope=17.0, model_slope_n=1000,
    observed_slope_n=10000
):
    alpha, P = reverse_probs(
        observed_slope, mean_slope_factor, mean_upper, mean_lower,
        min_slope=min_slope, max_slope=max_slope, 
        model_slope_n=model_slope_n, observed_slope_n=observed_slope_n
    )
    log_max, log_lower, log_upper = fit_log_of_normal(
        alpha, P
    )
    return log_max, log_lower, log_upper

def reverse_probs(
    observed_slope, mean_slope_factor, mean_upper, mean_lower,
    min_slope=0.2, max_slope=17.0, model_slope_n=1000,
    forward_mle_pdf=_forward_mle_pdf
):
    X0, P0 = forward_mle_pdf.X, forward_mle_pdf.P
    slope_array = -1 * np.linspace(
        min_slope, max_slope, model_slope_n, endpoint=False
    )
    prob_list = []
    for test_slope in slope_array:
        X = -1 * test_slope * X0
        P = P0 / trapezoid(X, P0)
        prob = np.interp(-1 * observed_slope, X, P)
        prob_list.append(prob)
    return -1 * slope_array, -1 * np.array(prob_list)
```
