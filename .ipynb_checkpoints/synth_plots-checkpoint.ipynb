While not used in the paper, this notebook is used to generate plots that have been used in talks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, poisson, linregress, norm, skew
from scipy.integrate import cumulative_trapezoid, trapezoid
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import NullLocator
import math
import time
import random
import pickle as pkl
from craterdata import *
%matplotlib inline
```

```python
model_lambda=100
n_steps=5
D = 10**np.arange(-1, 3, 0.1)
synth_area=10000
pf_slope = -3.0
pf = loglog_linear_pf(N1=0.001, slope=pf_slope)

synth_list, synth_age = synth_data(
    model_lambda=model_lambda, n_steps=n_steps, 
    area=synth_area, pf=pf, dmin=1, dmax=10**3.5
)
for i in range(n_steps):
    fig = plt.figure(figsize=(14.5, 6))
    ax = fig.add_subplot(111)
    plot_cumulative_unbinned(
        synth_list[i], synth_area, do_formatting=True,
        d_min=1, color=cs[1]
    )
    plt.xlim([10**-0.05, 10**2.3])
    plt.ylim([2E-5, model_lambda / 100 * 4E-2])

    plt.plot(D, synth_age * 10**pf(np.log10(D)), color='black', lw=0.5)
    pfig = plt.gcf()
    file_base = 'figs/variable_N/full_function_'
    pfig.savefig(
        file_base + str(model_lambda) + '_' + str(i) + '.pdf', 
        bbox_inches='tight'
    )
```

```python
model_lambda=20
n_steps=5
D = 10**np.arange(-1, 3, 0.1)
pf_slope = -3.0
pf = loglog_linear_pf(N1=0.001, slope=pf_slope)

synth_list, synth_age = synth_data(
    model_lambda=model_lambda, n_steps=n_steps, 
    area=synth_area, pf=pf, dmin=1, dmax=10**3.5
)
fig = plt.figure(figsize=(8, 6))
for i in range(n_steps):
    ax = fig.add_subplot(161 + i)
    plot_cumulative_unbinned(
        synth_list[i], synth_area, do_formatting=False,
        d_min=1, color=cs[2], plot_lines=False, plot_points=True,
        plot_point_error_bars=True, ms=2
    )
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim([10**-0.05, 10**0.8])
    plt.ylim([2E-5, model_lambda / 100 * 2E-2])
    ax.xaxis.set_major_formatter(ScalarFormatter())
    ax.xaxis.set_minor_locator(NullLocator())
    plt.ylabel(rf'Cumulative Crater Density (km$^{{-2}}$)', size=14)
    plt.xlabel('D (km)', size=14)
    plt.xticks([1, 2, 4])
    if i > 0:
        plt.ylabel(None)
        plt.yticks([])

    plt.plot(D, synth_age * 10**pf(np.log10(D)), color='black', lw=0.5)
fig.subplots_adjust(hspace=0.2, wspace=0)
pfig = plt.gcf()
pfig.savefig('figs/MN16_test.pdf', bbox_inches='tight')
```
