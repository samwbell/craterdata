```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.gridspec as gridspec
from matplotlib.ticker import ScalarFormatter
import matplotlib.patches as patches
import math
from scipy.stats.mstats import gmean
from scipy.stats import gamma, poisson, linregress, beta, norm, lognorm
from scipy import optimize
from scipy import interpolate
import time
import random
import pickle as pkl
from craterdata import *
%matplotlib inline
```

```python
mi_array = 10**np.linspace(0, 20, 10000)
K1_a = 1.0
K1_o = 1.0
μ_a = 0.55
μ_o = 0.40
ν = 1 / 3
Y0_a = 1.5E7
Y0_o = 1E6
ρt_a = 3000
ρt_o = 3000
vi = 700
ρi = 2500
g = 3.71
ap_D = D(mi_array, K1_a, μ_a, ν, Y0_a, ρt_o, g, vi, ρi)
oma_D = D(mi_array, K1_o, μ_o, ν, Y0_o, ρt_a, g, vi, ρi)
X = 10**np.linspace(-2, 0, 10000)
Y = 10**(-3 - 2.5 * np.log10(X))

fig = plt.figure(figsize=(10, 3))

ax = fig.add_subplot(121)
plt.plot(ap_D / 1000, ap_D / oma_D)
plt.xscale('log')
plt.ylim([0, 3])
plt.xlim([10**-2.5, 10**0])

ax = fig.add_subplot(122)
plt.plot(X, Y)
plt.plot(X * oma_D / ap_D, Y)
plt.xscale('log')
plt.yscale('log')
```
