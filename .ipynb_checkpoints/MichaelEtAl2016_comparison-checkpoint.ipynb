The craterstats package is designed to make specific plots from a command line interface, loading data from file.  To wrap it into a function, some finagling is required.  First, we load a dummy Craterpdf object from a file with dummy data.  Then, we create a function, get_Craterpdf(), that modifies this Cratercount object to match the N, area, and dmin that the user inputs.  Finally, we wrap get_Craterpdf into an m16_age_pdf_wrapped() function that creates a Craterpdf object for any N, area, and dmin, then converts it to a RandomVariable object.  Here, we use the NCF and NPF as the chronology and production functions.

```python
cc = cst.Cratercount(filename='saved/Im2.diam')
pf = cst.Productionfn(
    cst.PATH + "config/functions.txt", 
    'Moon, Neukum et al. (2001)'
)
cf = cst.Chronologyfn(
    cst.PATH + "config/functions.txt", 
    'Moon, Neukum et al. (2001)'
)

def get_Craterpdf(N, area, dmin, dmax=1000, pf=pf, cf=cf, cc=cc):
    d_range = np.array([dmin, dmax])
    cc.diam = tuple(sorted(list(
        np.random.uniform(dmin, dmax, N)
    )))
    cc.area = area
    return cst.Craterpdf(pf, cf, cc, d_range)

def m16_age_pdf_wrapped(N, area, dmin, dmax=1000, pf=pf, cf=cf, cc=cc):
    cpdf = get_Craterpdf(
        N, area, dmin, dmax=dmax, pf=pf, cf=cf, cc=cc
    )
    median, low_median, high_median = tuple(cpdf.t([.5, .1586, .8413]))
    return RandomVariable(
        cpdf.ts, cpdf.pdf, skip_fitting=True, median=median, 
        low_median=low_median, high_median=high_median
    )
```

```python
def age_method_fig(
    N, area, dmin, xlim=[0, 4.5], upshift=1.1, ax=None,
    figsize=(5, 4), label_shift_x=0.3, label_shift_y=-0.4,
    font_size=8
):

    if ax is None:
        fig = plt.figure(figsize=figsize)
        ax = fig.add_subplot(111)

    age_rv = age_pdf(N, area, dmin)
    age_rv_m16 = m16_age_pdf(N, area, dmin)
    age_rv_m16_w = m16_age_pdf_wrapped(N, area, dmin)
    age_rv_m16_s = age_scaled_N1_pdf(N, area, dmin)
    
    X = np.linspace(0, 4.5, 20000)
    
    age_rv = age_rv.match_X(X)
    age_rv_m16 = age_rv_m16.match_X(X)
    age_rv_m16_w = age_rv_m16_w.match_X(X)
    age_rv_m16_s = age_rv_m16_s.match_X(X)
    
    med_str = r'Median $\pm$ 34.1%'
    log_str = r'Log(normal $\pm$ $\sigma$)'
    
    pdfs = [
        age_rv_m16, age_rv_m16_w, age_rv_m16_s,
        age_rv_m16, age_rv, age_rv
    ]
    error_bar_types = [
        'median', 'median', 'median',
        'log', 'median', 'log',
    ]
    colors = [
        'xkcd:orange', 'xkcd:orange', 'xkcd:orange',
        'crimson', 'xkcd:shamrock', 'xkcd:vivid blue'
    ]
    labels = [
        'Michael et al. (2016)\n' + med_str,
        'CraterStats2.0\n' + med_str,
        'N(1) scaled by ncf_inv()\n' + med_str,
        'N(1) scaled by ncf_inv()\n' + log_str,
        'Random variable math\n' + med_str,
        'Random variable math\n' + log_str,
    ]
    
    for i in range(len(pdfs)):
        pdfs[i].plot(
            error_bar_type=error_bar_types[i], label=True, 
            upshift=i * upshift, 
            xlim=xlim, ax=ax, unit='Ga',
            color=colors[i], force_label_side='right', 
            label_shift_x=label_shift_x, 
            label_shift_y=label_shift_y
        )
        plt.text(
            xlim[0], i * upshift + 0.3, labels[i], 
            size=font_size, color=colors[i], ha='left'
        )
    plt.xlabel('Age (Ga)')
    plt.title(
        'N=' + str(N) + ', Area=' + scientific_notation(area) + \
            'km$^2$' + ', d$_{min}$=' + str(dmin) + 'km'
    )
```

The N(1) scaled method produces the same results as the CraterStats2.0 code and our implementation of Michael et al. (2016)'s age PDF equation (Equation 8 in Michael et al. (2016)).  These methods produce a meaningfully different PDF from random variable math for ages older than ~3Ga, where the NCF is no longer linear.

```python
age_method_fig(
    10, 2500, 1.0, xlim=[1.6, 4.25],
    label_shift_x=0.28
)
```

For age PDFs entirely younger than ~3Ga, where the NCF is linear because it assumes a constant cratering rate, all these methods produce the same PDF, although the log and median methods will produce different values and error bars.

```python
age_method_fig(
    10, 10000, 1.0, xlim=[0, 2.3],
    label_shift_x=-0.05,
    label_shift_y=-0.2,
)
```
