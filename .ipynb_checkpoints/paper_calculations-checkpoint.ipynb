```python
# Shows the calculation of where the new NPF's slope becomes shallower than -2, placing it in the
# quasi-equilibrium saturation domain.

logD = np.arange(-2,3.2,0.01)
plt.plot(logD, np.gradient(np.log10(npf_new(10**logD)), logD))
plt.plot(logD, -2 * np.ones(logD.shape[0]))

f_line1 = interpolate.interp1d(logD, np.gradient(np.log10(npf_new(10**logD)), logD))
f_line2 = interpolate.interp1d(logD, -2* np.ones(logD.shape[0]))

def intersection_equation(x):
    return f_line1(x) - f_line2(x)
result1 = optimize.root(intersection_equation, 0, method='hybr')
result2 = optimize.root(intersection_equation, 1.5, method='hybr')
10**result1.x, 10**result2.x
```
