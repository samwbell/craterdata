This notebook is designed to show how to use the cratrcountr package to do crater age calculations for the Moon.

Here, we load the necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import ScalarFormatter
import math
import time
import random
import pickle as pkl
from craterdata import *
%matplotlib inline
```

For our example data, we will use the counts of the Apollo 15 landing site from Robbins et al. (2014) provided by Stuart Robbins.  They come from LROC NAC image M146959973L.  The diameters are in units of pixels, and an adjustment is made for pixel size.  The area is an estimated area of the image.  It is not properly buffered, and it does not represent a clearly identified geologic unit per se.  These values are uncertain and could be wrong.  These data are for example purposes only.

```python
cfactor = np.cos(math.pi * (26 + 11 / 60) / 180) * 0.626608880720610
rawd = pd.read_csv('data/M146959973L.txt', delimiter='\t') * cfactor
ds = np.array(sorted(rawd['D_wm'].tolist()))/1000
area = 4000 * 2215 * cfactor**2 / 1000000
```

Let's start by plotting these data on a cumulative unbinned plot.  From this plot, we can tell that there is an equilibrium saturation inflection point at around ~0.09km.  In order to calculate ages, we have to exclude saturated portions of the curve.

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)
plot_cumulative_unbinned(ds, area, color='blue', ax=ax)
pfig = plt.gcf()
pfig.savefig(
    '../figs/M146959973L_cumulative_unbinned.pdf', 
    bbox_inches='tight'
)
```

Our next task is to estimate where to set dmin.  This distribution is in equilibrium saturation, so it's a question of estimating where the equilibrium saturation inflection point lies.  Visually, we might estimate the inflection point at ~0.09km, since that's where the inflection point in the cumulative plot is, but we can learn more by plotting the slopes around this point.  Let's start by calculating bin edges that fall at the geometric means of the diamters of individual craters.

```python
edges = np.sqrt(ds[1:] * ds[:-1])
```

This enables us to plot slope PDFs for the 1 + 2 * n craters in a bin centered around any particular crater.  In this example, we set n=4.  When we do this, we find that the sharp inflection point at ~0.09km may be a bit of a statisticical anomaly exacerbated by the cumulative plot.  First, we calculate the slope PDFs for bins centered around D > 0.06km.  We exclude the last n + 6 craters.  This is because we do not trust a truncated window for the last few craters to be statistically unbiased because of the chance that the production function might not produce enough craters for there to be at least one crater larger than the window.  For each the craters in this end portion, we calculate the same slope from the untruncated Pareto distribution.  (This code runs in 17.3 seconds on a relatively new MacBook Pro.)

```python
t1 = time.time()
n = 4
i_min = np.where(ds > 0.06)[0][0]
i_max = edges.shape[0] - (n + 5)
i_range = range(i_min, i_max)
slope_rv_list = [
    slope_pdf(ds, dmin=edges[i - n - 1], dmax=edges[i + n])
    for i in i_range
]
end_rv = pareto_pdf(ds[i_max:], dmin=edges[i_max - 1])
slope_rv_list = slope_rv_list + (n + 6) * [end_rv]
d_list = list(ds[i_range]) + list(ds[i_max:])
t2 = time.time()
print(format_runtime(t2 - t1))
```

We can plot the first D > 0.09 PDF using the .plot() method of RandomVariable objects.

```python
rv_i = np.where(np.array(d_list) > 0.09)[0][0]
slope_rv_list[rv_i].plot(label=True)
print('D = ' + str(round(d_list[rv_i], 5)))
```

To see what kind of error bars these are, we can check the .kind field of the RandomVariable object.

```python
slope_rv_list[rv_i].kind
```

It's using the mean method.  To show the asymmetry, we can use the linear method instead.

```python
slope_rv_list[rv_i].plot(label=True, error_bar_type='linear')
```

The slope centered around the 0.0915km crater is very shallow, making it unlikely that the production function is sampled out to this range.  If the production function were sampled out to 0.09km, as it looks from the cumulative plot, then the slope should be intermediate between the saturated slope and the unsaturated slope, not on the shallow end of the saturated slope.  To this in context and to see what value of dmin might make more sense, we plot the slopes from the 2 * n + 1 crater window around each crater over the whole diameter range.  Here, the orange shows the slope of the NPF.

```python
logD = np.linspace(np.log10(0.02), np.log10(0.4), 10000)
logRho = npf_new_loglog(logD)
npf_slope = np.gradient(logRho, logD)

slope_vals = np.array([rv.val for rv in slope_rv_list])
slope_lower = np.array([rv.lower for rv in slope_rv_list])
slope_upper = np.array([rv.upper for rv in slope_rv_list])

plt.errorbar(
    d_list, slope_vals, yerr=[slope_lower, slope_upper],
    fmt='.', elinewidth=0.5
)
plt.plot(10**logD, -1 * npf_slope)
plt.xlim([0.06, 0.3])
plt.xlabel('Crater Diameter (km)', size=12)
plt.ylabel('Negative Slope α', size=12)
```

And then over a narrower range.

```python
slope_vals = np.array([rv.val for rv in slope_rv_list])
slope_lower = np.array([rv.lower for rv in slope_rv_list])
slope_upper = np.array([rv.upper for rv in slope_rv_list])

plt.errorbar(
    d_list, slope_vals, yerr=[slope_lower, slope_upper],
    fmt='.', elinewidth=0.5
)
plt.plot(10**logD, -1 * npf_slope)
plt.xlim([0.06, 0.3])
plt.xlabel('Crater Diameter (km)', size=12)
plt.ylabel('Negative Slope α', size=12)
```

This is using mean method error bars.  Switching to the linear method:

```python
slope_vals = np.array([rv.as_kind('linear').val for rv in slope_rv_list])
slope_lower = np.array([rv.as_kind('linear').lower for rv in slope_rv_list])
slope_upper = np.array([rv.as_kind('linear').upper for rv in slope_rv_list])

plt.errorbar(
    d_list, mean_slopes, yerr=[slope_lower, slope_upper],
    fmt='.', elinewidth=0.5
)
plt.plot(10**logD, -1 * npf_slope)
plt.xlim([0.07, 0.13])
plt.xlabel('Crater Diameter (km)', size=12)
plt.ylabel('Negative Slope α', size=12)
```

From here, it is a subjective decision where to choose dmin.  We choose dmin = 0.09762675 (the geometric mean between two different diameters).

```python
dmin = 0.09762675
```

Now, we show dmin on the plot as a dotted line.

```python
slope_vals = np.array([rv.as_kind('linear').val for rv in slope_rv_list])
slope_lower = np.array([rv.as_kind('linear').lower for rv in slope_rv_list])
slope_upper = np.array([rv.as_kind('linear').upper for rv in slope_rv_list])

plt.errorbar(
    d_list, mean_slopes, yerr=[slope_lower, slope_upper],
    fmt='.', elinewidth=0.5
)
plt.plot(10**logD, -1 * npf_slope)
plt.xlim([0.07, 0.13])
plt.xlabel('Crater Diameter (km)', size=12)
plt.ylabel('Negative Slope α', size=12)

plt.plot([dmin, dmin], [0.7, 5.7], 'k:', lw=1)
plt.ylim([0.7, 5.7])
```

From here, we calculate N.

```python
unsaturated_ds = ds[ds > dmin]
N = unsaturated_ds.shape[0]
N
```

Now, we simply plug N, area, and dmin into the age_pdf function.  It returns a RandomVariable object for the random variable age, which we plot using the .plot() method of the RandomVariable object.

```python
age_pdf(N, area, dmin).plot(label=True, error_bar_type='median')
```

Now, let's break down how age_pdf works by working through the age PDF calculation step by step.  

First, we set the pf (production function) and cf_inv (inverse chronology function) equal to npf_new and ncf_inv, the defaults for age_pdf.  The NPF (Neukum Production Function) and NCF (Neukum Chronology Function) represent the production and chronology functions that define the Neukum chronology system for the Moon (Neukum, 1983; Neukum et al., 2001).  The "New NPF," represented by npf_new, is an update (Neukum et al., 2001) to the Neukum (1983) production function.  The NCF is written as a function that takes an age and gives an N(1) value.  The inverse NCF, represented by ncf_inv, is the function that translates N(1) values to age.  The inverse function cannot be solved analytically, so ncf_inv implements a numerical inverse function of the NCF.

```python
pf = npf_new
cf_inv = ncf_inv
```

First, we calculate the likelihood PDF of the model crater count $\lambda$ using the true_error_pdf function, which returns a RandomVariable object.  The maximum likelihood value is N = 16, and the upper error bar is slightly larger than the lower error bar.

```python
lambda_rv = true_error_pdf(N)
lambda_rv.plot(
    label=True, xlim=[5, 40]
)
plt.xlabel(r'Model Crater Count $\lambda$')
```

Then, we apply a uniform $\lambda$ prior.  This does not change the PDF, but we can show it visually by changing the error bar method to the median method.

```python
lambda_rv = lambda_rv.as_kind('median')
lambda_rv.plot(
    label=True, xlim=[5, 40]
)
plt.xlabel(r'Model Crater Count $\lambda$')
```

Next, we convert the $\lambda$ PDF to a crater density PDF by dividing by area.

```python
crater_density_rv = lambda_rv / area
crater_density_rv.plot(
    label=True, xlim=[2, 14], unit=r'$km^{-2}$',
    label_shift_x=-1.0
)
plt.xlabel(r'Crater Density (craters/km$^2$)')
```

Then, we use the production function (the NPF) to project our crater density PDF to an N(1) PDF.

```python
N1_shift = pf(1) / pf(dmin)
N1_rv = crater_density_rv * N1_shift
N1_rv.plot(
    label=True, xlim=[5.3E-4, 4.2E-3],
    unit=r'$km^{-2}$',
)
plt.xlabel(r'N(1) (craters/km$^2$)')
```

Finally, we use the apply2rv function to apply the inverse production function (the inverse NCF) to the N(1) PDF to produce an age pdf.

```python
age_rv = apply2rv(N1_rv, cf_inv)
age_rv.plot(
    label=True, unit='Ga'
)
plt.xlabel('Age (Ga)')
```
