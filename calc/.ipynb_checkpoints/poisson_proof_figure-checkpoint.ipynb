```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.gridspec as gridspec
from matplotlib.ticker import ScalarFormatter, FuncFormatter
import math
from scipy.stats.mstats import gmean
from scipy.stats import gamma, poisson, linregress, beta, norm
from scipy import optimize
from scipy import interpolate
import time
import random
import pickle as pkl
from craterdata import *
%matplotlib inline
```

```python
synth_area = 100000
pf = npf_new_loglog
dmin = 10.0
dmax = 10**3.5
model_lambda_list = [2.5, 5, 20]
n_steps = 1000000
inc=0.001

synth_list_list = []
synth_age_list = []
if False:  
    for i in range(len(model_lambda_list)):
        synth_list, synth_age = synth_data(
            model_lambda=model_lambda_list[i], n_steps=n_steps, 
            inc=inc, area=synth_area, pf=pf, dmin=dmin, dmax=dmax,
            dmax_tolerance=1E-7
        )
        file_str = '_NPF_' + str(model_lambda_list[i]) + '.pkl'
        synth_list_list.append(synth_list)
        synth_age_list.append(synth_age)
        with open('saved/synth_list' + file_str, 'wb') as f:
            pkl.dump(synth_list, f)
        with open('saved/synth_age' + file_str, 'wb') as f:
            pkl.dump(synth_age, f)
else:
    for i in range(len(model_lambda_list)):
        file_str = '_NPF_' + str(model_lambda_list[i]) + '.pkl'
        with open('saved/synth_list' + file_str, 'rb') as f:
            synth_list_list.append(pkl.load(f))
        with open('saved/synth_age' + file_str, 'rb') as f:
            synth_age_list.append(pkl.load(f))
```

```python
def poisson_proof_plot(synth_list, model_lambda):
    N_array = np.array([ds.shape[0] for ds in synth_list])
    N_values, N_counts = np.unique(N_array, return_counts=True)
    N_probabilities = N_counts / len(synth_list)
    norm = poisson.pmf(N_values, model_lambda).max()
    plt.plot(
        N_values, poisson.pmf(N_values, model_lambda) / norm, '.', 
        markersize=15, color='cornflowerblue', label='Poisson Distribution'
    )
    plt.plot(
        N_values, N_probabilities / norm, 'k.', markersize=3,
        label='1,000,000 Synthetics'
    )
    plt.legend(loc='upper right', fontsize=9)
    plt.xlabel('Total number of craters (N)')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.yticks([])
    # plt.xlim([0, 55])
```
