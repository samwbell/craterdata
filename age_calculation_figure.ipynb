```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import math
import time
import random
import pickle as pkl
import scipy as sp
from scipy.integrate import cumulative_trapezoid
from craterdata import *
%matplotlib inline
```

```python
N = 10
area = 2500
area2 = 10000
dmin = 1
pf = npf_new
cf_inv = ncf_inv

lambda_rv = true_error_pdf(N)
crater_density_rv = lambda_rv / area
N1_shift = pf(1) / pf(dmin)
N1_rv = crater_density_rv * N1_shift
age_rv = apply2rv(N1_rv, cf_inv)
m16_age_rv = N1_rv.scale(cf_inv)

def revive_box():
    ax.spines['top'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.spines['right'].set_visible(True)
    plt.yticks([])

def config_plot(
    xlim, ylim=[-0.07, 1.5], letter='a', 
    label='Step 1: Poisson PDF', xlabel=r'$\lambda$',
    letter_y_shift=0
):
    plt.xlabel(xlabel, labelpad=0)
    plt.ylim(ylim)
    plt.title(label, size=10)
    revive_box()
    plt.text(
        xlim[0] + 0.02 * (xlim[1] - xlim[0]), 
        ylim[1] - 0.1 + letter_y_shift, 
        letter, ha='left', va='top', size=20
    )

def poisson_PDF_step(
    letter='a', label='Step 1: Poisson PDF', kind='median', c=cs[1],
    lambda_rv=lambda_rv, letter_y_shift=0
):
    xlim = [3, 23]
    lambda_rv.plot(
        label=True, xlim=xlim, error_bar_type=kind,
        color=c, label_y=1.0,
        force_label_side='right'
    )
    config_plot(
        xlim, letter=letter, label=label, 
        xlabel=r'$\lambda$', letter_y_shift=letter_y_shift
    )

def crater_density_step(
    letter='c', label='Step 3: Divide by Area', kind='median', c=cs[4],
    crater_density_rv=crater_density_rv
):
    xlim = [1E-3, 9E-3]
    crater_density_rv.plot(
        label=True, xlim=xlim, unit=r'km$^{-2}$',
        rounding_n=2, color=c, label_y=1.0,
        force_label_side='right', error_bar_type=kind
    )
    config_plot(
        xlim, letter=letter, label=label, 
        xlabel=r'Crater Density (km$^{-2}$)'
    )

def N1_step(
    letter='d', label='Step 4: Calculate N(1)', kind='median', c=cs[4],
    N1_rv=N1_rv
):
    xlim=[1E-3, 9E-3]
    N1_rv.plot(
        label=True, rounding_n=2, xlim=xlim,
        unit=r'km$^{-2}$', color=c, label_y=1.0,
        force_label_side='right', error_bar_type=kind
    )
    config_plot(
        xlim, letter=letter, label=label, 
        xlabel=r'N(1) (km$^{-2}$)'
    )
    
def age_step(
    letter='e', label='Step 5: Calculate Age', age_rv=age_rv, 
    kind='median', c=cs[7]
):
    xlim=[2.9, 3.7]
    age_rv.plot(
        label=True, unit='Ga',
        xlim=xlim, color=c, label_y=1.0,
        force_label_side='right', error_bar_type=kind
    )
    config_plot(
        xlim, letter=letter, label=label, 
        xlabel='Age (Ga)'
    )

lkind = 'median'

fig = plt.figure(figsize=(11.3, 9.0))
gs = gridspec.GridSpec(5, 5, figure=fig)

ax = plt.subplot(gs[0, 0])
poisson_PDF_step(
    letter='a', label='Step 1: Poisson PDF', kind=lkind, c=cs[1]
)
plt.ylabel('Uniform\n' + r'$\lambda$ Prior', rotation=0, ha='right', va='center')

ax = plt.subplot(gs[0, 1])
poisson_PDF_step(
    letter='b', label='Step 2: Apply to Prior', c=cs[2]
)

ax = plt.subplot(gs[0, 2])
crater_density_step(
    letter='c', label='Step 3: Divide by Area', c=cs[8]
)

ax = plt.subplot(gs[0, 3])
N1_step(
    letter='d', label=r'Step 4: Multiply by $\phi$', c=cs[11]
)

ax = plt.subplot(gs[0, 4])
age_step(
    letter='e', label='Step 5: Apply cf_inv', c=cs[5]
)

ax = plt.subplot(gs[1, 0])
poisson_PDF_step(
    letter='f', label='Step 1: Poisson PDF', kind=lkind, c=cs[1]
)
plt.ylabel('Uniform\nAge Prior', rotation=0, ha='right', va='center')

ax = plt.subplot(gs[1, 1])
crater_density_step(
    letter='g', label='Step 2: Divide by Area', kind=lkind, c=cs[7]
)

ax = plt.subplot(gs[1, 2])
N1_step(
    letter='h', label=r'Step 3: Multiply by $\phi$', kind='linear', c=cs[10]
)

ax = plt.subplot(gs[1, 3])
age_step(
    letter='i', label='Step 4: Apply cf_inv', age_rv=m16_age_rv, 
    kind=lkind, c=cs[4]
)

ax = plt.subplot(gs[1, 4])
age_step(
    letter='j', label='Step 5: Apply to Prior', age_rv=m16_age_rv, c=cs[5]
)

ax = plt.subplot(gs[2, 0])
age_step(
    letter='k', label='Age PDF With Age Prior', age_rv=m16_age_rv, c=cs[5]
)
plt.ylabel(
    r'$\lambda$' + ' PDF of\nUniform\nAge Prior', 
    rotation=0, ha='right', va='center'
)

ax = plt.subplot(gs[2, 1])
N1_ap_rv = m16_age_rv.apply(ncf)
N1_step(
    letter='l', label='Apply CF', c=cs[11], N1_rv=N1_ap_rv
)

ax = plt.subplot(gs[2, 2])
crater_density_ap_rv = N1_ap_rv / N1_shift
crater_density_step(
    letter='m', label=r'Divide by $\phi$', c=cs[8],
    crater_density_rv=crater_density_ap_rv
)

ax = plt.subplot(gs[2, 3])
lambda_ap_rv = crater_density_ap_rv * area
poisson_PDF_step(
    letter='n', label=r'Multiply by Area', c=cs[1],
    lambda_rv=lambda_ap_rv
)

ax = plt.subplot(gs[2, 4])
X = np.linspace(0, 4, 10000)
P = 0.0 * X + 1.0
uniform_rv = RandomVariable(X, P, kind='median')
N1_ua_rv = uniform_rv.apply(ncf)
cd_ua_rv = N1_ua_rv / N1_shift
lambda_ua_rv = cd_ua_rv * area
poisson_PDF_step(
    letter='o', label=r'Prior in $\lambda$ Space', c=cs[0],
    lambda_rv=lambda_ua_rv, letter_y_shift=0.1
)

ax = plt.subplot(gs[3, 0])
poisson_PDF_step(
    letter='p', label='Step 1: Poisson PDF', kind=lkind, c=cs[1]
)
plt.ylabel(
    'Uniform\n' + r'$\lambda$ Prior' + '\nWith NPF\nError Model', 
    rotation=0, ha='right', va='center'
)

ax = plt.subplot(gs[3, 1])
poisson_PDF_step(
    letter='q', label='Step 2: Apply to Prior', c=cs[2]
)

ax = plt.subplot(gs[3, 2])
crater_density_step(
    letter='r', label='Step 3: Divide by Area', c=cs[8]
)

ax = plt.subplot(gs[3, 3])
X = np.linspace(0, 2, 10000)
P = norm.pdf(X, 1, npf_error(0.5))
N1_shift_rv = RandomVariable(X, P)
N1_pfe_rv = crater_density_rv * N1_shift_rv
N1_step(
    letter='s', label=r'Step 4: Multiply by $\phi$', c=cs[11],
    N1_rv=N1_pfe_rv
)

ax = plt.subplot(gs[3, 4])
age_pfe_rv = N1_pfe_rv.apply(ncf_inv)
age_step(
    letter='t', label='Step 5: Apply cf_inv', c=cs[5],
    age_rv=age_pfe_rv
)

ax = plt.subplot(gs[4, 0])
poisson_PDF_step(
    letter='u', label='Step 1: Poisson PDF', kind='linear', c=cs[1]
)
plt.ylabel(
    'Uniform\n' + r'$\lambda$ Prior' + '\nRecommended\nError Bars', 
    rotation=0, ha='right', va='center'
)

ax = plt.subplot(gs[4, 1])
poisson_PDF_step(
    letter='v', label='Step 2: Apply to Prior', c=cs[2]
)

ax = plt.subplot(gs[4, 2])
crater_density_step(
    letter='w', label='Step 3: Divide by Area', c=cs[8]
)

ax = plt.subplot(gs[4, 3])
N1_step(
    letter='x', label=r'Step 4: Multiply by $\phi$', c=cs[11]
)

ax = plt.subplot(gs[4, 4])
age_step(
    letter='y', label='Step 5: Apply cf_inv', c=cs[5]
)


plt.subplots_adjust(wspace=0, hspace=1.0)
pfig = plt.gcf()
pfig.savefig('figs/age_example.pdf', bbox_inches='tight')
```
