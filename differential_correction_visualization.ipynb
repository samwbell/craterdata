```python
# This visualizes the Michael (2013) differential plot
# correction as a function of cumulative slope. The 
# solid line represents 100/decade bins, the dashed 
# line represents 18/decade "Neukum" bins, and the 
# dotted line represents sqrt(2) bins.

w = 2**neukum_bwe
fig = plt.figure(figsize=(4, 2))
ax = fig.add_subplot(111)
m = np.arange(-5.0 ,-0.01, 0.01)
ax.plot(m, differential_correction(w, m), 'k--')
ax.plot(m, differential_correction(10**(1 / 100), m), 'k')
ax.plot(m, differential_correction(np.sqrt(2), m), 'k:')
plt.xlabel('Cumulative Slope', size=12)
plt.ylabel('Correction Factor', size=12)
plt.xlim([-4, -1])
plt.ylim([0.925, 1.005])
pfig = plt.gcf()
pfig.savefig(
    'figs/differential_correction_visualization.pdf',
    bbox_inches='tight'
)
```
